# Can a linear trend trick you into thinking there is a phase offset in your signal?

- YES.
- This is where I dump my thoughts and sanity checks

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 1000)
sin = np.sin(t)

islope = 0.5*t+0.0
dslope = -0.5*t+0.0

peaksin = np.argmax(sin)
plt.figure()
print('Time of first peak without slop:', t[peaksin])
plt.title('Sinusoid no slope')
plt.plot(t, sin, label='simple sinuisoid')
plt.plot(t, islope, label='increasing slope')
plt.plot(t, dslope, label='decreasing slope')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.xlim(np.min(t), np.max(t))
plt.axvline(x=t[peaksin],color='grey', label='Time of first peak')
plt.legend()

slope = 0.5*t+0.0
peakarg = np.argmax((sin+islope)[:500])
plt.figure()
print('Time of first peak with increaing slope:', t[peakarg])
plt.title('Sinusoid with slope')
plt.plot(t,sin+islope, label='sin + increasing slope')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.xlim(np.min(t), np.max(t))
plt.axvline(x=t[peakarg], color='grey',label='Time of  first peak')
plt.legend()

peakarg = np.argmax((sin+dslope)[:500])
plt.figure()
print('Time of first peak with increaing slope:', t[peakarg])
plt.title('Sinusoid with slope')
plt.plot(t,sin+dslope, label='sin + decreasing slope')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.xlim(np.min(t), np.max(t))
plt.axvline(x=t[peakarg], color='grey',label='Time of  first peak')
plt.legend()
```
